```python
# Import required libraries and dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(market_data_df)
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(
    market_data_scaled,
    columns = market_data_df.columns,
    index= market_data_df.index
)

# Copy the crypto names from the original data


# Set the coinid column as index


# Display sample data
df_market_data_scaled.head()
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_vals = list(range(1,11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
for k in k_vals:
    Kmeans = KMeans(n_clusters=k, random_state=42)
    Kmeans.fit(df_market_data_scaled)
    inertia.append(Kmeans.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_vals, 
    "inertia": inertia,
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.plot.line(x="k",
                   y="inertia",
                   title="Elbow Curve",
                   xticks=k_vals)
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:** 4

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
clusters = model.predict(df_market_data_scaled)

# View the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the DataFrame
df_market_data_scaled_clustered = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scaled_clustered['cluster'] = clusters

# Display sample data
df_market_data_scaled_clustered.head()
```

```python
# Create a scatter plot using Pandas plot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Use "rainbow" for the color to better visualize the data.
plt.figure(figsize=(10, 8))
plt.scatter(
    df_market_data_scaled_clustered["price_change_percentage_24h"],
    df_market_data_scaled_clustered["price_change_percentage_7d"],
    c=df_market_data_scaled_clustered["cluster"],
    cmap="rainbow"
)
plt.colorbar(label="Cluster")
plt.xlabel("Price Change Percentage (24h)")
plt.ylabel("Price Change Percentage (7d)")
plt.title("Cryptocurrency Clusters")
plt.grid(True)
plt.show()
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
crypto_data = pca.fit_transform(df_market_data_scaled)
# View the first five rows of the DataFrame. 
print(crypto_data[:5])
```

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
print(f"variance / component: {explained_variance}")

total_explained_variance = sum(explained_variance)
print(f"total variance: {total_explained_variance}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The three principal components capture approximately 89.5% of the total variance in the original data. This indicates that we've preserved most of the information while reducing dimensions from seven to three.

```python
# Create a new DataFrame with the PCA data.
pca_crypto_df = pd.DataFrame(
    crypto_data,
    columns=['PC1', 'PC2', 'PC3'],
    index= market_data_df.index)

# Creating a DataFrame with the PCA data


# Copy the crypto names from the original data


# Set the coinid column as index


# Display sample data
pca_crypto_df.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_values = list(range(1,11))
# Create an empty list to store the inertia values
pca_inertia = []


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using PCA DataFrame.
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans_pca = KMeans(n_clusters=k, random_state=42)
    kmeans_pca.fit(pca_crypto_df)
    pca_inertia.append(kmeans_pca.inertia_)
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {
    'k': k_values,
    'inertia': pca_inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)

# Display the DataFrame
df_elbow_pca
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.plot.line(x="k",
                   y="inertia",
                   title="Elbow Curve",
                   xticks=k_values)
```

#### Answer the following questions: 
* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** This is the same as the best value found using the original data, which confirms that our dimensionality reduction preserved the underlying cluster structure of the data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
kmeans_pca = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(pca_crypto_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_clusters = kmeans_pca.predict(pca_crypto_df)

# Print the resulting array of cluster values.
print(pca_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_pca_crypto_clusters = pca_crypto_df.copy()

# Add a new column to the DataFrame with the predicted clusters
df_pca_crypto_clusters['cluster'] = pca_clusters

# Display sample data
df_pca_crypto_clusters.head()
```

```python
# Create a scatter plot by setting `x="PCA1"` and `y="PCA2"`. 
plt.scatter(
    df_pca_crypto_clusters["PC1"],
    df_pca_crypto_clusters["PC2"],
    c=df_pca_crypto_clusters["cluster"],
    cmap="rainbow"
)
```

### Determine the Weights of Each Feature on each Principal Component

```python
# Use the columns from the original scaled DataFrame as the index.
df_feature_weights = pd.DataFrame(
    pca.components_.T,
    columns=["PC1", "PC2", "PC3"],
    index=market_data_df.columns
)

df_feature_weights
```

#### Answer the following question: 

* **Question:** Which features have the strongest positive or negative influence on each component? 
 
* **Answer:** PC1 is most strongly influenced by long-term price changes (200d and 1y), indicating this component primarily captures long-term performance trends. PC2 is heavily influenced by short-to-medium term price changes (7d, 14d, 24h, and 30d), suggesting this component represents recent price momentum. PC3 shows strong influence from medium-term changes (30d and 60d), possibly capturing cyclical price patterns.


    

```python
plt.figure(figsize=(8, 6))
df_feature_weights.plot(kind='bar', figsize=(12, 8))
plt.title('Feature Weights for Principal Components')
plt.xlabel('Features')
plt.ylabel('Weight')
plt.grid(True)
```
